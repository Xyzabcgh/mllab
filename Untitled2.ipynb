```python
class Node:
    def __init__(self,data,fval,level):
        self.data=data
        self.fval=fval
        self.level=level
    def generate_child(self):
        x,y=self.find(self.data,'_')
        val_list=[[x+1,y],[x-1,y],[x,y+1],[x,y-1]]
        children=[]
        for i in val_list:
            child=self.shuffle(self.data,x,y,i[0],i[1])
            if child is not None:
                childnode=Node(child,self.level+1,0)
                children.append(childnode)
        return children
    def find(self,puz,x):
        for i in range(0,len(puz)):
            for j in range(0,len(puz)):
                if puz[i][j]==x:
                    return i,j
    def shuffle(self,puz,x,y,x1,y1):
        if x1>=0 and x1<len(puz) and y1>=0 and y1<len(puz):
            tmp=[]
            tmp=self.copy(puz)
            tmpd=tmp[x1][y1]
            tmp[x1][y1]=tmp[x][y]
            tmp[x][y]=tmpd
            return tmp
        else:
            return None
    def copy(self,puz):
        tmp=[]
        for i in puz:
            t=[]
            for j in i:
                t.append(j)
            tmp.append(t)
        return tmp

class Puzzle:
    def __init__(self,n):
        self.n=n
        self.open=[]
        self.close=[]
    def accept(self):
        puz=[]
        for i in range(0,self.n):
            tmp=input().split(" ")
            puz.append(tmp)
        return puz
    def f(self,start,goal):
        return self.h(start,goal)+start.level
    def h(self,start,goal):
        tmp=0
        for i in range(0,self.n):
            for j in range(0,self.n):
                if start.data[i][j]!=goal.data[i][j] and start.data[i][j]!='_':
                    tmp=tmp+1
        return tmp
    def process(self):
        print("Enter start state")
        start=self.accept()
        print("Enter goal state")
        goal=self.accept()
        start=Node(start,0,0)
        goal=Node(goal,0,0)
        start.fval=self.f(start,goal)
        self.open.append(start)
        while True:
            cur=self.open[0]
            print("-----------------------------------------------------")
            for i in cur.data:
                for j in i:
                    print(j,end=" ")
                print("")
            print("-----------------------------------------------------")
            if self.h(cur,goal)==0:
                break
            for i in cur.generate_child():
                i.fval=self.f(i,goal)
                self.open.append(i)
            self.close.append(cur)
            del self.open[0]
            self.open.sort(key=lambda x:x.fval,reverse=False)
            
puz = Puzzle(3)
puz.process() 
                
        
        
        
```

```python
#kmeans 
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

class Kmeans:
    def __init__(self,no_clusters,no_iter=1000):
        self.no_clusters=no_clusters
        self.no_iter=no_iter
        
    def fit(self,X):
        self.centroids=X[np.random.choice(X.shape[0],self.no_clusters,replace=False)]
        for _ in range(self.no_iter):
            self.labels=self.assign_labels(X)
            self.new_centroid=self.generate_centroid(X,self.labels)
            if np.all(self.centroids==self.new_centroid):
                break
            self.centroids=self.new_centroid
    def assign_labels(self,X):
        return np.argmin(np.linalg.norm(X[:,np.newaxis]-self.centroids,axis=2),axis=1)
    def generate_centroid(self,X,labels):
        return np.array([X[labels==i].mean(axis=0) for i in range(self.no_clusters)])
X,_=make_blobs(300,centers=3,random_state=123)
k=Kmeans(3)
k.fit(X)
#labels=k.assign_labels(X)
print("Cluster Assignments:", k.labels)
print("Final Centroids:", k.centroids)

plt.scatter(X[:,0],X[:,1],c=k.labels)
plt.scatter(k.centroids[:,0],k.centroids[:,1],c='red',marker='x',label='centroid')
plt.title("Kmeans Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
